````python
def twoSum(nums, target):
    seen = {}
    ret = []
    for i, val in enumerate(nums):
        rem = target - val
        if rem in seen:
            ret.append([i, seen.get(rem)])
        else:
            seen[nums[i]] = i
    return ret

print(twoSum([1,-4, 4, -1], 0))
````

````python
class Solution:
    def evaluate(self, s: str, knowledge: List[List[str]]) -> str:

        key_value = {x[0]: x[1] for x in knowledge }
        result = []
        i = 0
        while i <len(s):
            if s[i] == '(':
                start = i
                while i<len(s) and s[i] != ')':
                    i +=1
                key = s[start+1:i]
                i+=1 # go past )
                result.append(key_value.get(key, '?'))
            else:
                result.append(s[i])
                i+=1
        return "".join(result)
    
````

# contextlib 
- helps you create your own context managers easily.
- Below code can be rewritten using "with"
```
f = open("data.txt")
try:
    data = f.read()
finally:
    f.close
```
as
```
with open("data.txt") as f:
    data = f.read()
```

````python
from contextlib import contextmanager

@contextmanager
def myopen(filename, mode="r"):
    f = open(filename, mode)
    try:
        yield f         # this becomes the value of "as f"
    finally:
        f.close()  

with myopen("data.txt") as f:
    print(f.read())
````

````python
#class based
class myopen:
    def __init__(self, filename, mode="r"):
        self.filename = filename
        self.mode = mode
        self.file = None

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_value, traceback):
        self.file.close()

with myopen("data.txt") as f:
    print(f.read())
````

````python
from contextlib import contextmanager
import atexit

@contextmanager
def log_time(task_name):
    start = time.time()
    print(f"[{task_name}] started...")
    yield
    end = time.time()
    print(f"[{task_name}] completed in {end - start:.2f} seconds")

with log_time("process_data"):
    time.sleep(2)

with log_time("save_results"):
    time.sleep(1)

@atexit.register
def goodbye():
    print("Program is exiting. Cleaning up...") # test in script. run in normal exit not crash

print("Main program is running...")
````

````python
def detect_repeats(items):
    first_index = {}
    messages = []
    template = "{elem} repeats at index {i}, first found at index {j}"

    for i, item in enumerate(items):
        # setdefault adds the value and returns the value or if key existed it returns previous value
        j = first_index.setdefault(item, i)
        if j != i:
            messages.append(template.format(elem=item, i=i, j=j))

    return messages

````

````python
from functools import reduce
# Taking the input
user_input = input().strip().lower().split()

operation = user_input[0]
numbers = list(map(float, user_input[1:]))
print(numbers)
print(operation)


# Write your code below
def perform_operation(lst, name):
    if len(lst) == 0: return "No numbers provided"
    if name == "sum":
        return reduce(lambda x, y : x + y, lst)
    elif name == "product":
        return reduce(lambda x, y : x * y, lst)
    elif name == "average":
        return reduce(lambda x, y : (x + y)/2, lst)
    else:
        return "Invalid operation specified"

perform_operation(numbers, operation )
````
